```python
import matplotlib.pyplot as plt
```

```python
#plot histogram of sentence length
def plot_sentence_length_histogram(df):
    df['sentence_length'] = df['SENTENCE'].apply(lambda x: len(x.split()))
    df['sentence_length'].hist(bins=100)
    plt.title('Sentence Length Histogram')
    #add a line to the plot for the average sentence length
    plt.axvline(df['sentence_length'].mean(), color='k', linestyle='dashed', linewidth=1)
    #add a line to the plot for the highest frequency
    plt.axvline(df['sentence_length'].mode()[0], color='r', linestyle='dashed', linewidth=1)
    #x axis values per 10
    plt.xticks(range(0, max(df['sentence_length']), 10))
    plt.xlabel('Sentence Length')
    plt.ylabel('Frequency')
    plt.show()
```

```python
plot_sentence_length_histogram(df)
```

```python
# add wordclouds
from wordcloud import WordCloud
def plot_word_cloud(df, title):
    wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = None, 
                min_font_size = 10).generate(' '.join(df['SENTENCE']))
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 
    plt.title(title)
    plt.show()
```

```python
plot_word_cloud(df_train, 'Training Data Word Cloud')
```

```python
plot_word_cloud(df_val, 'Validation Data Word Cloud')
```

```python
plot_word_cloud(df_test, 'Test Data Word Cloud')
```
